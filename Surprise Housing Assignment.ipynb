```python
from sklearn.model_selection import train_test_split
Y=df['SalePrice']
X=df.drop(columns=['SalePrice'],axis=1)
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
df.columns.tolist()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# X_train,X_test,Y_train,Y_test
# Train the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predict the target variable on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)

# Get the coefficients and corresponding feature names
coefficients = model.coef_
feature_names = X.columns

# Create a DataFrame to store the coefficients and feature names
coefficients_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
coefficients_df['Abs_Coefficient'] = np.abs(coefficients_df['Coefficient'])  # Absolute values for easier interpretation
coefficients_df = coefficients_df.sort_values(by='Abs_Coefficient', ascending=False)

# Print the significant variables
significant_variables = coefficients_df[coefficients_df['Abs_Coefficient'] > 0]
print("Significant variables in predicting SalePrice:")
print(significant_variables)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error




# Define the range of alpha values to search
alphas = [ 0.001, 0.01, 0.1, 1, 10, 100,200,300,500,600,700,800,900,1000,10000]

# Ridge regression
ridge_params = {'alpha': alphas}
ridge_grid = GridSearchCV(Ridge(), param_grid=ridge_params, scoring='neg_mean_squared_error', cv=5)
ridge_grid.fit(X_train, Y_train)
ridge_best_alpha = ridge_grid.best_params_['alpha']

# Lasso regression
lasso_params = {'alpha': alphas}
lasso_grid = GridSearchCV(Lasso(), param_grid=lasso_params, scoring='neg_mean_squared_error', cv=5)
lasso_grid.fit(X_train, Y_train)
lasso_best_alpha = lasso_grid.best_params_['alpha']

# Train the models with the best alphas
ridge_model = Ridge(alpha=ridge_best_alpha)
ridge_model.fit(X_train, Y_train)

lasso_model = Lasso(alpha=lasso_best_alpha)
lasso_model.fit(X_train, Y_train)

# Evaluate the models
ridge_pred = ridge_model.predict(X_test)
lasso_pred = lasso_model.predict(X_test)

ridge_rmse = mean_squared_error(Y_test, ridge_pred, squared=False)
lasso_rmse = mean_squared_error(Y_test, lasso_pred, squared=False)

print("Optimal lambda for Ridge regression:", ridge_best_alpha)
print("Optimal lambda for Lasso regression:", lasso_best_alpha)
print("Ridge RMSE:", ridge_rmse)
print("Lasso RMSE:", lasso_rmse)
```
